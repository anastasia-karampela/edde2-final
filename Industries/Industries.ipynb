```
# Εισαγωγή βιβλιοθηκών

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import statistics
import datetime
```

```
# Δημιουργία DataFrames
# Eισαγωγή νέας στήλης

df_amazon = pd.read_csv('amazon.com.csv')
df_amazon.insert(0, 'Name', 'Amazon.com')

df_amazon.head()
```

```
df_apple = pd.read_csv('apple_inc.csv')
df_apple.insert(0, 'Name', 'Apple Inc')

df_apple.head()
```

```
df_carlsberg = pd.read_csv('carlsberg.csv')
df_carlsberg.insert(0, 'Name', 'Carlsberg')

df_carlsberg.head()
```

```
df_lufthansa = pd.read_csv('deutsche_lufthansa.csv')
df_lufthansa.insert(0, 'Name', 'Deutsche Lufthansa')

df_lufthansa.head()
```

```
df_eurobank = pd.read_csv('eurobank_ergasias_services_and_holdings.csv')
df_eurobank.insert(0, 'Name', 'Eurobank Ergasias Services and Holdings')

df_eurobank.head()
```

```
df_johnson = pd.read_csv('johnson&johnson.csv')
df_johnson.insert(0, 'Name', 'Johnson & Johnson')

df_johnson.head()
```

```
df_microsoft = pd.read_csv('microsoft_corporation.csv')
df_microsoft.insert(0, 'Name', 'Microsoft Corporation')

df_microsoft.head()
```

```
df_national_bank = pd.read_csv('national_bank_of_greece.csv')
df_national_bank.insert(0, 'Name', 'National Bank of Greece')

df_national_bank.head()
```

```
df_reckitt_benckiser = pd.read_csv('reckitt_benckiser_group.csv')
df_reckitt_benckiser.insert(0, 'Name', 'Reckitt Benckiser Group')

df_reckitt_benckiser.head()
```

```
df_coca_cola = pd.read_csv('the_coca_cola_company.csv')
df_coca_cola.insert(0, 'Name', 'The Coca Cola Company')

df_coca_cola.head()
```

```
df_zoom = pd.read_csv('zoom_video_communications.csv')
df_zoom.insert(0, 'Name', 'Zoom Video Communications')

df_zoom.head()
```

```
# Ένωση των DataFrames

df_industries = pd.concat([df_amazon, df_apple, df_carlsberg, df_coca_cola, df_eurobank, df_johnson, df_lufthansa, df_microsoft, df_national_bank, df_reckitt_benckiser, df_zoom])
print(df_industries) 
```

```
# Καθαρισμός των δεδομένων

df_industries.isnull().sum()
missing_values = ["n/a", "na", "NaN", "--", 999, 9999]   
df_industries.replace(to_replace = missing_values, value = np.nan, inplace = True)
df_industries.dropna(inplace=True) 
df_industries
```

```
# Μετατροπή της ημερομηνίας από object σε datetime σε νέα στήλη

print(df_industries.dtypes)
df_industries['Datetime'] = pd.to_datetime(df_industries['Date'], format='%Y-%m-%d')
print(df_industries.dtypes)
df_industries.head()
```

```
# Οπτικοποίηση των δεδομένων 

import matplotlib.dates as dates 
groups = df_industries.groupby('Name')

for name, group in groups:
  fig, ax = plt.subplots(figsize=(7,4), linewidth=4)
  ax.plot(group['Datetime'], group['Close'], label=name)
  ax.legend([name])
  ax.xaxis.set_major_locator(dates.MonthLocator())
  ax.xaxis.set_major_formatter(dates.DateFormatter('%b'))
  plt.title('Αξία μετοχών: Ιούνιος 2019 - Μάιος 2020')
  plt.ylabel('Αξία μετοχής')
  plt.xlabel('Μήνες')

  plt.show()
```
